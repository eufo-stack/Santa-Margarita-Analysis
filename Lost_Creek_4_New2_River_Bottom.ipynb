Monthly Avg per Hr

```python
# monthly avg per hr bot
def get_month_yr(datee):
    mon = datee.strftime('%m')
    yr = datee.year
    hr = datee.strftime('%H')
    return str(yr) + '-' + str(mon) + '-' + str(hr)
bot['yr-month-hr'] = bot['Unnamed: 1'][1:].apply(get_month_yr)

bot_cut = bot.iloc[1:8259].drop(['Unnamed: 3','Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6','Unnamed: 7', 'Unnamed: 8'],axis=1)    #keeping indexes the same
bot_cut
```

```python
# monthly avg
bot_cut['Unnamed: 2'] = bot_cut['Unnamed: 2'].apply(pd.to_numeric)
avgs = bot_cut.groupby('yr-month-hr')['Unnamed: 2'].mean().reset_index()
avgs
```

```python
from datetime import datetime
lst = []
for a in np.arange(len(avgs['yr-month-hr'])):
    datetime_object = datetime.strptime(avgs['yr-month-hr'][a], '%Y-%m-%H')
    lst.append(datetime_object)
avgs['time'] = lst
avgs
```

```python
#new bot
def get_hr(datee):
    hr = datee.strftime('%H')
    return str(hr)
avgs['hr'] = avgs['time'].apply(get_hr)
avgs
#
```

```python
#new bot
high = avgs[avgs['hr'] == '14']
plt.plot(high['yr-month-hr'],high['Unnamed: 2'])
plt.xticks(rotation=90)
plt.title("New 4 River Bottom Monthly Avg Hr 14")
plt.xlabel("Year-Month-Hour")
plt.ylabel("Avg Temp (C)")
plt.show()

low = avgs[avgs['hr'] == '06']
plt.plot(low['yr-month-hr'],low['Unnamed: 2'])
plt.xticks(rotation=90)
plt.title("New 4 River Bottom Monthly Avg Hr 6")
plt.xlabel("Year-Month-Hour")
plt.ylabel("Avg Temp (C)")
plt.ylim([0, 40])
plt.show()
#
```

```python
# monthly avg per hr top
def get_month_yr(datee):
    mon = datee.strftime('%m')
    yr = datee.year
    hr = datee.strftime('%H')
    return str(yr) + '-' + str(mon) + '-' + str(hr)
top['yr-month-hr'] = top['Unnamed: 1'][1:].apply(get_month_yr)

top_cut = top.iloc[1:8259].drop(['Unnamed: 3','Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6','Unnamed: 7'],axis=1)    #keeping indexes the same
# monthly avg
top_cut['Unnamed: 2'] = top_cut['Unnamed: 2'].apply(pd.to_numeric)
avgstop = top_cut.groupby('yr-month-hr')['Unnamed: 2'].mean().reset_index()

lsttop = []
for b in np.arange(len(avgstop['yr-month-hr'])):
    datetime_object = datetime.strptime(avgstop['yr-month-hr'][b], '%Y-%m-%H')
    lsttop.append(datetime_object)
avgstop['time'] = lsttop

def get_hr(datee):
    hr = datee.strftime('%H')
    return str(hr)
avgstop['hr'] = avgstop['time'].apply(get_hr)

hightop = avgstop[avgstop['hr'] == '14']
plt.plot(hightop['yr-month-hr'],hightop['Unnamed: 2'])
plt.xticks(rotation=90)
plt.title(title + " Monthly Avg Hr 14")
plt.xlabel("Year-Month-Hour")
plt.ylabel("Avg Temp (C)")
plt.ylim([0, 40])
plt.show()

lowtop = avgstop[avgstop['hr'] == '06']
plt.plot(lowtop['yr-month-hr'],lowtop['Unnamed: 2'])
plt.xticks(rotation=90)
plt.title(title + " Monthly Avg Hr 6")
plt.xlabel("Year-Month-Hour")
plt.ylabel("Avg Temp (C)")
plt.ylim([0, 40])
plt.show()
```
