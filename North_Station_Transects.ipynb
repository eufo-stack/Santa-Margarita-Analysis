Monthly Avg per Hr

```python
# monthly avg per hr bot
def get_month_yr(datee):
    mon = datee.strftime('%m')
    yr = datee.year
    hr = datee.strftime('%H')
    return str(yr) + '-' + str(mon) + '-' + str(hr)
bot['yr-month-hr'] = bot['Unnamed: 1'][1:].apply(get_month_yr)

bot_cut = bot.iloc[1:8259].drop(['Unnamed: 3','Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6','Unnamed: 7'],axis=1)    #keeping indexes the same
bot_cut
```

```python
# monthly avg
bot_cut['Unnamed: 2'] = bot_cut['Unnamed: 2'].apply(pd.to_numeric)
#avgs = bot_cut.groupby('yr-month-hr')['Unnamed: 2'].mean().reset_index()
avgs = bot_cut
avgs
```

```python
#new bot
def get_hr(datee):
    hr = datee.strftime('%H')
    return str(hr)
def get_mon(datee):
    mon = datee.strftime('%m')
    return int(mon)
avgs['hr'] = avgs['Unnamed: 1'].apply(get_hr)
avgs['month'] = avgs['Unnamed: 1'].apply(get_mon)
avgs
#
```

```python
# plot hour by temp to find highest/lowest avg for hour
summer1 = avgs[avgs['month'] > 3]
summer2 = summer1[summer1['month'] < 10]
winter1 = avgs[avgs['month'] <= 3]
winter2 = avgs[avgs['month'] >= 10]
winter3 = winter2.append(winter1)
plt.scatter(summer2['hr'],summer2['Unnamed: 2'])
plt.title("Lost Creek 3 (Apr - Sept) Monthly Avg Per Hour")
plt.xlabel("Hour")
plt.ylabel("Avg Temp (C)")
plt.show()
plt.scatter(winter3['hr'],winter3['Unnamed: 2'])
plt.title("Lost Creek 3 (Nov - Mar) Monthly Avg Per Hour")
plt.xlabel("Hour")
plt.ylabel("Avg Temp (C)")
plt.show()
```

```python
# summer: 13, 5
# winter: 14, 6
# separate bot plot by time of year, tmin/tmax
high1 = summer2[summer2['hr'] == '13']
plt.plot(high1['Unnamed: 1'],high1['Unnamed: 2'])
plt.xticks(rotation=90)
plt.title("Lost Creek 3 (296.30 El.) Monthly (Apr - Sept) Avg Hr 13")
plt.xlabel("Month")
plt.ylabel("Avg Temp (C)")
plt.ylim([-10, 50])
plt.show()

low1 = summer2[summer2['hr'] == '05']
plt.plot(low1['Unnamed: 1'],low1['Unnamed: 2'])
plt.xticks(rotation=90)
plt.title("Lost Creek 3 (296.30 El.) Monthly (Apr - Sept) Avg Hr 5")
plt.xlabel("Month")
plt.ylabel("Avg Temp (C)")
plt.ylim([-10, 50])
plt.show()

high2 = winter3[winter3['hr'] == '14']
plt.plot(high2['Unnamed: 1'],high2['Unnamed: 2'])
plt.xticks(rotation=90)
plt.title("Lost Creek 3 (296.30 El.) Monthly (Nov - Mar) Avg Hr 14")
plt.xlabel("Month")
plt.ylabel("Avg Temp (C)")
plt.ylim([-10, 50])
plt.show()

low2 = winter3[winter3['hr'] == '06']
plt.plot(low2['Unnamed: 1'],low2['Unnamed: 2'])
plt.xticks(rotation=90)
plt.title("Lost Creek 3 (296.30 El.) Monthly (Nov - Mar) Avg Hr 6")
plt.xlabel("Month")
plt.ylabel("Avg Temp (C)")
plt.ylim([-10, 50])
plt.show()

# together
low2botall = low1.append(low2)
plt.plot(low2botall['Unnamed: 1'],low2botall['Unnamed: 2'])
plt.xticks(rotation=90)
plt.title("North Station 11 (248.80 El.) Daily Min Temp (Apr - Sept) Hr 5, (Nov - Mar) Hr 6")
plt.xlabel("Month")
plt.ylabel("Avg Temp (C)")
plt.ylim([-10, 50])
plt.grid()
plt.show()

high2botall = high1.append(high2)
plt.plot(high2botall['Unnamed: 1'],high2botall['Unnamed: 2'])
plt.xticks(rotation=90)
plt.title("North Station 11 (248.80 El.) Daily Max Temp (Apr - Sept) Hr 13, (Nov - Mar) Hr 14")
plt.xlabel("Month")
plt.ylabel("Avg Temp (C)")
plt.ylim([-10, 50])
plt.grid()
plt.show()
```
