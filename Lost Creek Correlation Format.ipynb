Lost Creek Correlation Formatting

```python
# imports
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
# function for cleaning up each file
def clean(df):
    df.dropna(subset = ['Unnamed: 2'], inplace=True)
    start = datetime.datetime(2020, 4, 4, 8)
    end = datetime.datetime(2021, 3, 14, 11)
    df.drop(0,inplace=True)
    df = df[df['Unnamed: 1'] > start]
    df = df[df['Unnamed: 1'] < end]
    return df
```

```python
# creating a base dataframe for each of the Lost Creek sites
import datetime
df = pd.read_excel('20706412_run1_LC3.xlsx')
df = clean(df)
df = df.drop({'Plot Title: 20706412_run1','Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 7'},1)
df = df.rename({'Unnamed: 1': 'Date','Unnamed: 2':'Lost Creek 3 (296.30 El.)'}, axis='columns')
df
```

```python
# adding the other Lost Creek sites
inputs = ['20706565_run1_LC1.xlsx','20706413_run1_LC4.xlsx','20706562_run1_LC2.xlsx','20706560_run2_LC5.xlsx']
names = ['New 1 Lost Creek (339.11 El.)','Lost Creek 4 (317.99 El.)','Lost Creek 2 (317.15 El.)','Lost Creek 5 (313.96 El.)']
for i in np.arange(4):
    df_add = pd.read_excel(inputs[i])
    df_add = clean(df_add)
    print(df_add['Unnamed: 2'])
    df[names[i]] = df_add.loc[:,'Unnamed: 2']
    
```
